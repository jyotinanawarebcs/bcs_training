```python
import pandas as pd
```

```python

df1 = pd.DataFrame({
    'id': [1, 2, 3],
    'name': ['Karan', 'Rohan', 'Meena'],
    'age': [25, 30, 22]
})

df2 = pd.DataFrame({
    'id': [1, 2, 4],
    'city': ['Mumbai', 'Pune', 'Delhi']
})
```

```python
print(df1)
```

```python
print(df2)
```

```python
merged = pd.merge(df1,df2,on="id")
```

```python
print(merged)
```

```python

data = {
    "Name": ["Amit", "Riya", "John"],
    "Age": [25, 30, 22],
    "City": ["Delhi", "Mumbai", "Pune"]
}
df = pd.DataFrame(data)
print(df)
```
